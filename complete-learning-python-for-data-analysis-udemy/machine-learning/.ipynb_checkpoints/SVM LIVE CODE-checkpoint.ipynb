```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets
```

```python
# review SVMs from Andrew Ng's course
iris = datasets.load_iris()

X = iris.data
Y = iris.target
```

```python
print iris.DESCR
```

```python
from sklearn.svm import SVC
```

```python
# initialize SVC object
model = SVC()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
# split data into training + testing set
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.4,
                                                 random_state=3)
```

```python
# fit model to the data
model.fit(X_train,Y_train)
```

```python
from sklearn import metrics
```

```python
# what model predicts based on X test data
predicted = model.predict(X_test)

expected = Y_test
```

```python
# model predicts Y correctly about 97% of the time
print metrics.accuracy_score(expected,predicted)
```

```python
from sklearn import svm
```

```python
# looking at kernel methods so not worrying about test set
# select only 2 features
X = iris.data[:,:2]
Y = iris.target
```

```python
C = 1.0
```

```python
# creating and fitting linear SVM model
svc = svm.SVC(kernel='linear',C=C).fit(X,Y)
```

```python
# Gaussian radial based function
rbf_svc = svm.SVC(kernel='rbf',gamma=0.7,C=C).fit(X,Y)
```

```python
# 3rd degree polynomial function
poly_svc = svm.SVC(kernel='poly',degree=3,C=C).fit(X,Y)
```

```python
# linear SVC model
lin_svc = svm.LinearSVC(C=C).fit(X,Y)
```

```python
h = 0.02

# grab min and max of first feature
x_min = X[:,0].min() - 1
x_max = X[:,0].max() + 1

# grab min and max of target
y_min = X[:,1].min() - 1
y_max = X[:,1].max() + 1
```

```python
# create meshgrid to arange x and y min and max
xx,yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
```

```python
xx
```

```python
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
```

```python
for i,clf in enumerate((svc,lin_svc,rbf_svc,poly_svc)):
    
    plt.figure(figsize=(15,15))
    
    # four figures, last argument is location
    plt.subplot(2,2,i+1)
    
    # adjust space in between plots
    plt.subplots_adjust(wspace=0.4,hspace=0.4)
    
    # change data shape and predict
    Z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    
    Z = Z.reshape(xx.shape)
    
    # set color
    plt.contourf(xx,yy,Z,cmap=plt.cm.terrain,alpha=0.5,linewidths = 0)
    
    # scatterplot of training set
    plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Dark2)
    
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])
```
